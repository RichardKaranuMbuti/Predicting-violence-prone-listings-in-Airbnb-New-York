About the dataset
The dataset contains Arbnb listings for New York.The last column 'Violence reported ' is a simulated and therefore it is important to note that the type of violence indicated didnt happen and wasnt reported.

Impoting the dataset

```python
import pandas as pd
df=pd.read_csv("D:\Datasets\AB_NYC_2019.csv")
df.head(5)
```

When running the model I want to map the column for violence reported as follows
1-Gender
2-Race
3-Other
4-No violence reported

```python
vrp = {1: "Gender",2: 'Race',3:'Other',4:'None'}
df.Violence_reported = [vrp[item] for item in df.Violence_reported]
```

Next is select the target variables that can be used to make predictions.This involves selecting the new target variables and storing them in a new dataframe.The selected target variables to be used to train the model are

Host id-uniquence idetifies each host
Neighbourhood group- will be useful in decision tree classification
number of reviews-The assumption is that if a client is exposed to any kind of violence they are most likely to make a review
reviews per month
Violence reported

```python
df = df[[ 'id','neighbourhood_group','number_of_reviews','reviews_per_month','Violence_reported']]
df
```

The next step is encoding the neighbourhood group column as machine learning algorithms only work with numbers

```python
nbgr = {'Brooklyn': 1,'Manhattan': 2,'Queens':3,'Staten Island':4,'Bronx':5}
df.neighbourhood_group = [nbgr[item] for item in df.neighbourhood_group]
df
```

```python
df.shape
```

Next we can drop any row wth any missing values and compare the structure of the dataframe to what it was when missing values were included 

```python
df=df.dropna()
df.shape
```

The next thing is to split the dataset into input set and output sets.Input set refers to what the user will enter so as to enable the model make predictions. The output set refers to what the model is expected to predict
To create the input set we drop the predicted column
The output set will be created by mapping the predicted column into a new dataframe object

```python
#Inputset
X=df.drop(columns=['Violence_reported'])
X
```

```python
#Output set
y=df['Violence_reported']
y
```

There are many machine learning algorithms but I will settle on Decision trees to avoid overfitting the model and because it has higher prediction accuracy

```python
from sklearn.tree import DecisionTreeClassifier

```

The next step is fitting the model into a new instance called model and feed both the input set and output sets

```python
model=DecisionTreeClassifier()
model.fit(X,y)
```

```python
X
```

After fitting the model we can now use it to make predictions.
We just need to include the input set in the same order it was in the input dataframe we used to fit the model.That is

id,neighbourhood_group,number_of_reviews,reviews_per_month and then store the output in a new variable predictions

```python
predictions=model.predict([[2539,1,9,0.21]])
predictions
```

From the above output we can see the model predicts a Racial abuse from the used input set
